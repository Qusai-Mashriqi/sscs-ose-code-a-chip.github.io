```python
# @title [VLSI'26] Interactive Hardware Generation for Drone PWM Controllers
# @markdown **Author:** Qusai Mashriqi (Senior Student, Yarmouk University)
# @markdown **Topic:** Educational RTL Design & Visualization for Drone Systems (NASAQ)

'''
Licensed under the Apache License, Version 2.0 (the "License");
you may not use this file except in compliance with the License.
'''

import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, HTML

# 1. Introduction Header
display(HTML("""
<div style="background-color: #f0f7ff; padding: 20px; border-radius: 10px; border-left: 5px solid #00629B;">
    <h2 style="color: #00629B;">Project Overview</h2>
    <p>This notebook demonstrates the design of a <b>Pulse Width Modulation (PWM)</b> generator, a core hardware component in flight controllers.
    As part of the <b>NASAQ</b> framework for drone traffic management, this module translates digital commands into physical signals for motor speed control.</p>
    <p><b>Target:</b> Open-source ASIC flow (Sky130 PDK).</p>
</div>
"""))

# 2. Verilog RTL Description
print("\n" + "="*50)
print("STEP 1: Verilog RTL Representation")
print("="*50)

verilog_code = """
module pwm_gen (
    input clk,           // System Clock
    input [7:0] duty,    // 8-bit Duty Cycle (0-255)
    output reg pwm_out   // PWM Signal to ESC
);
    reg [7:0] counter = 0;

    always @(posedge clk) begin
        counter <= counter + 1;
        // Comparator logic for PWM generation
        pwm_out <= (counter < duty);
    end
endmodule
"""
print(verilog_code)

# 3. Interactive Python Simulator
display(HTML("<h2 style='color: #00629B;'>STEP 2: Signal Simulator</h2>"))

def simulate_pwm(duty_cycle):
    period = 256
    time_steps = list(range(period))
    signal = [1 if t < duty_cycle else 0 for t in time_steps]

    plt.figure(figsize=(12, 4))
    plt.step(time_steps, signal, where='post', color='#00629B', linewidth=2.5)
    plt.fill_between(time_steps, signal, step="post", alpha=0.15, color='#00629B')
    plt.ylim(-0.2, 1.2)
    plt.xlim(0, 255)
    plt.title(f"Hardware-Level Signal: Duty Cycle = {duty_cycle}/255", fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

# Run once with a static value for GitHub preview
simulate_pwm(128)

# Interactive slider (will work in Colab, but won't break GitHub metadata)
try:
    widgets.interact(simulate_pwm, duty_cycle=(0, 255, 1))
except:
    print("Run in a live kernel (like Colab) to use the interactive slider.")

# 4. ASIC Flow Path
display(HTML("""
<div style="margin-top: 30px; padding: 15px; background-color: #fff4e5; border-radius: 5px;">
    <h3>Step 3: Path to Silicon</h3>
    <p>This RTL is compatible with <b>OpenLane</b>. The synthesis flow involves:</p>
    <ul>
        <li><b>Yosys:</b> Translating Verilog to logic gates.</li>
        <li><b>OpenROAD:</b> Placement and Routing on Sky130 layers.</li>
        <li><b>Magic/KLayout:</b> DRC/LVS checks and GDSII generation.</li>
    </ul>
</div>
"""))
```
